```python
# define your error/cost function

def costFunction(x_list,y_list,m, b):
    error = 0
    for i in range(0,x_list.shape[0]):
        error += (y_list[i] - (m*x_list[i] + b))**2
    return error/(float(x_list.shape[0]))
```

```python
def gradientDescent(m,b,x_list,y_list,learning_rate):
    b_grad = 0
    m_grad = 0
    N = float(x_list.shape[0])
    error = []
    
    for i in range(x_list.shape[0]):
        b_grad += -x_list[i] * (y_list[i] - (m*x_list[i] + b))
        m_grad += -(y_list[i] - (m*x_list[i] + b))
    b_grad = b_grad *(2.0/N)
    m_grad = m_grad *(2.0/N)
    new_b = b - (learning_rate*b_grad)
    new_m = m - (learning_rate*m_grad)
    return [new_m,new_b]
    
    
```
